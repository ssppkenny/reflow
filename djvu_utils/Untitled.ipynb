```python
import mydjvulib
import numpy as np
from PIL import Image

def get_image(pageno, filepath):
    bs, w, h = mydjvulib.get_page(pageno, filepath)
    arr = np.frombuffer(bs, dtype=np.uint8).reshape((h,w))
    img = Image.fromarray(arr, "L")
    return img
```

```python
for i in range(100):
    get_image(9, "dvurog1.djvu")
```
